```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.DataFrame({'x':[4,8,13,7],'y':[11,4,5,14]})
data

x_mean=np.mean(data['x'])
y_mean=np.mean(data['y'])

data_cent=data-[x_mean,y_mean]

cov_mat=np.cov(data_cent.T)
cov_mat

eigvalue,eigvector=np.linalg.eig(cov_mat)

print("Eigen values:\n",eigvalue)
print("Eigen vector:\n",eigvector)

principle_comp=eigvector[:,np.argmax(eigvalue)]

data_projected=data_cent.dot(principle_comp)
print(f"\nProjected Data (1D):\n{data_projected}")

plt.scatter(data['x'], data['y'], color='blue', label='Original Data',s=100)
plt.scatter(data_projected, np.zeros_like(data_projected),color='red', label='Projected Data',s=100,marker='x')
plt.xlabel('Feature x')
plt.ylabel('Feature y')
plt.title('PCA: Dimensionality Reduction from 2D to 1D')
plt.legend()
plt.show()
```
